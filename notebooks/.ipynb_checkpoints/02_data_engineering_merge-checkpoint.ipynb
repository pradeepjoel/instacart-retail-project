```python
import pandas as pd
```

```python
order_products = pd.read_csv(
    "../data_raw/order_products__train.csv"
)

products = pd.read_csv("../data_raw/products.csv")
aisles = pd.read_csv("../data_raw/aisles.csv")
departments = pd.read_csv("../data_raw/departments.csv")

print("Data loaded")
```

```python
products_full = products.merge(aisles, on="aisle_id", how="left") \
                        .merge(departments, on="department_id", how="left")

transactions = order_products.merge(products_full, on="product_id", how="left")
```
