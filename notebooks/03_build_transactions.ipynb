Imports + load processed fact table

```python
import os
import pandas as pd

PROJECT_ROOT = os.path.abspath("..")
PROCESSED_DIR = os.path.join(PROJECT_ROOT, "data", "processed")

fact_order_items_df = pd.read_parquet(os.path.join(PROCESSED_DIR, "fact_order_items.parquet"))
print("Loaded:", fact_order_items_df.shape)
```

Basic sanity checks

```python
min_item_freq = 2000
min_basket = 3
max_basket = 50

item_counts = fact_order_items_df["product_id"].value_counts()
keep_items = item_counts[item_counts >= min_item_freq].index

filtered_df = fact_order_items_df[fact_order_items_df["product_id"].isin(keep_items)].copy()

basket_sizes = filtered_df.groupby("order_id").size()
keep_orders = basket_sizes[(basket_sizes >= min_basket) & (basket_sizes <= max_basket)].index
filtered_df = filtered_df[filtered_df["order_id"].isin(keep_orders)]

print("Unique orders:", filtered_df["order_id"].nunique())
print("Unique products:", filtered_df["product_id"].nunique())
print("Rows:", filtered_df.shape[0])
```
