```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("QualityChecks") \
    .getOrCreate()

transactions = spark.read.parquet("../data_clean/transactions")
```

### Perform Basic Data Quality Checks

This cell runs simple validation checks on the cleaned transaction dataset.

- Total row count confirms dataset size
- Non-null `order_id` count ensures valid transactions
- Distinct `product_id` count shows how many unique products are present

These checks help confirm data completeness and consistency before enrichment.


```python
transactions.count()
transactions.select("order_id").dropna().count()
transactions.select("product_id").distinct().count()
```
